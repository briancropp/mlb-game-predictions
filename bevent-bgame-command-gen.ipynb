```python
import pandas as pd # Import pandas dataframe
import os # Import os package
import glob # Import glob package
import re # Import regular expressions
```

```python
'''
Make list of event files for further printing of microsoft command line action
'''

data_list = glob.glob("data/not-clean/event-files-1518/*.EV*") # Create list of all event data files
for i in range(len(data_list)): # For loop to parse correct file names of event files
    data_list[i] = data_list[i][32:52] # Parse data_list for file names
data_list
```

```python
'''
Print Microsoft Command Line actions for bevent.exe
'''
for event in data_list: # For loop to print the nessecary commands for the bevent retrosheet program
    print('bevent -y '+ event[0:4]+" "+event+" > "+event[0:7]+"_bevent.txt")
```
